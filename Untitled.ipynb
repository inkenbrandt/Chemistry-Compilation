```python
testdata = 'H:/Google Drive/WORK/Groundwater Chemistry/AsSDWIS.txt'
AsData = pd.read_csv(testdata)

def unitfix(x):
    if x[0] == 'MG/L':
        return x[1]*1000.0
    else:
        return x[1]
    
def iscen(x):
    if x[0] <= x[1]:
        return False, np.nan
    else:
        return True, x[0]

AsData['CONC'] = AsData[['UOM_CODE','CONC_MSR']].apply(lambda x: unitfix(x), 1)
AsData['DET_CONC'] = AsData[['DET_UOM','DET_LIM']].apply(lambda x: unitfix(x), 1)
AsData['CENS'] = AsData[['CONC','DET_CONC']].apply(lambda x: iscen(x)[0], 1)
AsData['CONC'] = AsData[['CONC','DET_CONC']].apply(lambda x: iscen(x)[1], 1)
AsData = AsData[(AsData.CONC < 50.0)]
from lifelines.datasets import load_lcd
from lifelines import KaplanMeierFitter


T = AsData['CONC']
C = AsData['CENS'] #boolean array, True if observed.

kmf = KaplanMeierFitter()
kmf.fit(T,C, left_censorship=True)
kmf.cumulative_density_
kmf.plot() #will plot the CDF
plt.xlim(0,50)

xa = AsData['CONC']
xa.plot.hist(bins=120,normed=True, cumulative=True)
```
